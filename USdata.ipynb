```python
concated=pd.concat([df_filtered1,df_filtered2,df_filtered3,df_filtered4,df_filtered5,df_filtered6])
```

```python
concated
```

```python
import pandas as pd

# Step 1: 각 (년월, 물품명) 조합이 몇 개인지 세기
count_df = concated.groupby(['년월', '물품명']).size().reset_index(name='count')

# Step 2: count가 2인 것만 필터링
valid_items = count_df[count_df['count'] == 2][['년월', '물품명']]

# Step 3: 원래 데이터에서 조건에 맞는 행만 필터링
filtered_df = concated.merge(valid_items, on=['년월', '물품명'], how='inner')

# Step 4: (년월, 물품명)별로 금액 합산
result_df = filtered_df.groupby(['년월', '물품명'], as_index=False)['금액'].sum()
```

```python
result_df
```

```python
# 분기 열 생성: '년월'이 YYYYMM 형식이므로 datetime으로 변환 후 분기 추출
result_df['분기'] = pd.to_datetime(result_df['년월'], format='%Y%m').dt.to_period('Q')

# 물품명과 분기별 평균 금액 계산
quarterly_avg = result_df.groupby(['물품명', '분기'])['금액'].mean().reset_index()

# 보기 좋게 정렬
quarterly_avg = quarterly_avg.sort_values(by=['물품명', '분기'])
```

```python
quarterly_avg
```

```python
quarter_counts = quarterly_avg.groupby('물품명')['분기'].nunique()

# 12개 분기 모두 존재하는 품목만 필터링
valid_items = quarter_counts[quarter_counts == 12].index

# 해당 품목만 포함된 데이터프레임 생성
filted = quarterly_avg[quarterly_avg['물품명'].isin(valid_items)]
```

```python
filted
```

```python
filted['분기별 상승률'] = (
    filted.groupby('물품명')['금액']
    .transform(lambda x: x.pct_change())
)
```

```python
filted
```

```python
pivot_df = filted.pivot(index='물품명', columns='분기', values='분기별 상승률')
```

```python
pivot_df
```

```python
pivot_df = pivot_df.reset_index()
```

```python
import numpy as np

# '2022Q1'을 제외한 분기 컬럼만 선택
cols_to_check = pivot_df.iloc[:,2:].columns
mask = (pivot_df[cols_to_check].abs() > 5).any(axis=1)
pivot_df=pivot_df[~mask]
```

```python
pivot_df = pivot_df.reset_index()
```

```python
from sklearn.preprocessing import StandardScaler

# 수치형 변화율 열만 선택 (첫 번째 열은 품목명이라 제외)
value_columns = pivot_df.columns[3:]

# 스케일러로 z-score 정규화
scaler = StandardScaler()
pivot_df[value_columns] = scaler.fit_transform(pivot_df[value_columns])
```

```python
pivot_df
```

```python
import torch
feature_tensor = torch.tensor(pivot_df.iloc[:,3:].values, dtype=torch.float64)
```

```python
from sklearn.cluster import KMeans
print("Any NaN:", torch.isnan(feature_tensor).any().item())
print("Any Inf:", torch.isinf(feature_tensor).any().item())
print("Max:", feature_tensor.max().item())
print("Min:", feature_tensor.min().item())
```

```python
common_params = {
    "n_init": "auto",
    "random_state": 170,
}
y_pred1 = KMeans(n_clusters=10, **common_params).fit_predict(feature_tensor)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertias = []
K_range = range(1, 15)
for k in K_range:
    kmeans = KMeans(n_clusters=k, **common_params)
    kmeans.fit(feature_tensor)
    inertias.append(kmeans.inertia_)

plt.plot(K_range, inertias, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

scores = []
K_range = range(2, 15)  # silhouette
for k in K_range:
    labels = KMeans(n_clusters=k, **common_params).fit_predict(feature_tensor)
    score = silhouette_score(feature_tensor, labels)
    scores.append(score)

plt.plot(K_range, scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method')
plt.show()
```
